# Assignment 07: Support vector machines

In this assignment, we will build support vector machine models using both linear and nonlinear kernels. Recall that support vector machines aim to maximize the margin between the two classes by finding the "cleanest separating hyperplane" between the two classes. When the two classes are not separable by a hyperplane, we use slack variables which allow, but also penalize points that fall to the wrong side of the separating hyperplace. The scale of the penalty is a parameter to tune.

## Lets make it faster

```python
import multiprocessing
from datetime import datetime
```

```python
def svm_worker(c_val):
    print "done\t", '%.e' %(c_val), "\t", datetime.now().strftime("%H:%M:%S")
    # create clf
    svm_clf = svm.SVC(kernel='linear', C=c_val)
    
    # fit svm clf
    svm_clf.fit(titanic_train[input_cols], titanic_train['Survived'])
    
    # make predictions
    train_preds = svm_clf.predict(titanic_train[input_cols])
    holdo_preds = svm_clf.predict(titanic_holdo[input_cols])

    # calculate accuracies
    train_accur = np.mean(titanic_train['Survived'] == train_preds)
    holdo_accur = np.mean(titanic_holdo['Survived'] == holdo_preds)

    print "done\t", '%.e' %(c_val), "\t", datetime.now().strftime("%H:%M:%S")
    return c_val, train_accur, holdo_accur
```

```python
c_vals = 10**np.linspace(-6,6,13)
p = multiprocessing.Pool(8)
results = p.map(svm_worker, c_vals)
```

```python
tuning_accuracies_df = pd.DataFrame(results, columns=["c","train","holdo"])
tuning_accuracies_df.set_index("c", inplace=True)

tuning_accuracies_df.holdo.plot()
tuning_accuracies_df.train.plot()
plt.xscale('log')
plt.legend()
```
